```python
conversion_rates = {"PKR": 0.0036, "INR": 0.011,"CNY": 0.14, "AED": 0.27}
```

```python
data["Launched.Price.Pakistan.USD"] = data["Launched.Price.Pakistan.PKR"] * conversion_rates["PKR"]
data["Launched.Price.India.USD"] = data["Launched.Price.India.INR"] * conversion_rates["INR"]
data["Launched.Price.China.USD"] = data["Launched.Price.China.CNY"] * conversion_rates["CNY"]
data["Launched.Price.Dubai.USD"] = data["Launched.Price.Dubai.AED"] * conversion_rates["AED"]
```

```python
def categorize_price(price):
    if price < 300:
        return "Budget"
    elif 300 <= price <= 700:
        return "Mid-Range"
    else:
        return "Premium"
```

```python
data["Price.Category.USA"] = data["Launched.Price.USA.USD"].apply(categorize_price)

sns.countplot(data=data, x="Company.Name", hue="Price.Category.USA")
plt.title("Smartphone Brands by Price Category")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.legend(title="Price Category")
plt.xticks(rotation=45)
plt.show()
```
