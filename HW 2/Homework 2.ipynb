```python
plt.hist(data2['PM2.5_AQI_Value'],  bins=30, edgecolor='black')
plt.xlabel('PM2.5 AQI Values')
plt.ylabel('Count')
plt.title('Histogram of PM2.5 AQI Values')
plt.show()
```

```python
sns.kdeplot(data = data1, x='Mortality_Rate', fill=True)
plt.xlabel('Mortality Rate')
plt.title('Density Plot of the lung Cancer Martality Rate')
plt.show()
```

```python
# Create the figure with white background
plt.figure(figsize=(10, 6))

# Set the background color of the axes (where the plot is)
plt.gca().set_facecolor('lightgrey')

# Add only horizontal gridlines (with zorder to ensure gridlines stay in the background)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5, color='white', zorder=0)

# Plot the histogram with lines around bins (with zorder to bring it to the front)
plt.hist(data2['PM2.5_AQI_Value'], bins=30, density=True, alpha=0.6, color='blue', edgecolor='black', linewidth=1.2, label='Histogram', zorder=3)

# Plot the density plot (KDE) with color below the curve (with zorder to bring it to the front)
sns.kdeplot(data2['PM2.5_AQI_Value'], color='red', linewidth=2.5, fill=True, alpha=0.4, label='Density Plot (KDE)', zorder=3)

# Add labels and title with bold font
plt.title('PM2.5 AQI Distribution with Density Overlay', fontsize=16, fontweight="bold")
plt.xlabel('PM2.5 AQI Value', fontsize=14, fontweight="bold")
plt.ylabel('Density', fontsize=14, fontweight="bold")
plt.legend(fontsize=12)

# Remove the y-axis line and ticks
plt.gca().spines['left'].set_visible(False)
plt.gca().tick_params(axis='y', length=0)

# Remove top, right, and bottom borders
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

# Add the descriptive line at the bottom left
plt.figtext(0.05, -0.08, "This plot represents the distribution of PM2.5 AQI values with a density overlay.", 
            ha="left", fontsize=12, color="black")

# Show the plot
plt.tight_layout()  # Ensure the text doesn't overlap with the plot
plt.show()
```
